```python
url = 'https://climaterightsdatabase.com'

import requests
from bs4 import BeautifulSoup

def find_all_links(url):
    # Fetch the webpage
    response = requests.get(url)
    response.raise_for_status()  # Raises an exception for HTTP errors

    # Parse the webpage with BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all links and filter to ignore non-http links
    links = [link['href'] for link in soup.find_all('a', href=True) if link['href'].startswith('http')]
    return links

def fetch_html_from_links(links):
    x=0

    pages_content = []
    for link in links:
        try:
            # Fetch each linked page
            response = requests.get(link)
            response.raise_for_status()  # Raises an exception for HTTP errors
            # Store the HTML content
            pages_content.append((link, response.text))
        except requests.RequestException as e:
            print(f"Failed to fetch {link}: {e}")
        x+=1
        if x ==20:
            break
        
    return pages_content

# The URL where you want to search links
url_to_search = 'https://climatecasechart.com/non-us-principle-law/united-kingdom/'

# Fetch all links from the homepage
links_found = find_all_links(url_to_search)

# Fetch the HTML from each found link
html_contents = fetch_html_from_links(links_found)

# You can now process or display the HTML content as needed
for link, content in html_contents:
    print(f"Content from {link}:")
    print(content[:500])  # Printing the first 500 characters of each HTML content
```
