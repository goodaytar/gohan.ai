```python
import pandas as pd
```

```python
path = 'data/climatecasechart/climatecasechart.csv'
df = pd.read_csv(path)
```

```python
# df.dropna(inplace=True)
principal_laws = df[['Principal Laws']].dropna()
principal_laws = principal_laws[principal_laws['Principal Laws'].str.contains('united kingdom', case=False)]
df_uk = df.iloc[principal_laws.index]
df_uk
# [df['Principal Laws'].str.contains('United Kingdom')]
# df_uk.shape
```

```python
def process(string):
    return ''.join(string.split('United Kingdom>')[1:]).lower().strip()

df_uk['Principal Laws Processed'] = df_uk['Principal Laws'].apply(process)
df_uk
```

```python
uk_laws = df_uk['Principal Laws Processed'].unique().tolist()
uk_laws_count = {i:0 for i in uk_laws}
for law in df_uk['Principal Laws Processed'].tolist():
    if law in uk_laws:
        uk_laws_count[law] += 1
# sort dict in descending order
uk_laws_count = dict(sorted(uk_laws_count.items(), key=lambda item: item[1], reverse=True))
uk_laws_count
```

```python
top_law = list(uk_laws_count.keys())[0]
df_uk[df_uk['Principal Laws Processed']==top_law]
```
